# Stack and unstack

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oo = pd.read_csv('data/olympics.csv',skiprows=4)
oo
```

### Plot the number of gold medals won by the USA male and female athletes throughout the history of the Olympics?

```python
ug = oo[(oo.NOC == 'USA') & (oo.Medal == 'Gold')]
ug.head()
```

```python
ug.groupby(['Edition','Gender']).size()
```

```python
ug.groupby(['Edition','Gender']).size().unstack('Gender')
```

```python
ug.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0)
```

```python
ug.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0).plot()
```

### Plot the 5 athletes who have won the most gold medals over the history of the Olympics. When there is a tie, consider the number of silver medals, then bronze medals.

```python
g = oo.groupby(['Athlete','Medal']).size()
```

```python
g
```

```python
g = g.unstack('Medal',fill_value=0)
g
```

```python
g = g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
g
```
