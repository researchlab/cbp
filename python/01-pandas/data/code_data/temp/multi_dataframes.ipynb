## Join

```python
noc.head()
```

```python
del noc['American Samoa*.1']
```

```python
noc.head()
```

```python
m3 = pd.merge(oo, noc, left_on='NOC',right_on='ASA')
m3.head()
```

```python
m3.tail()
```

```python
noc = noc.rename(index=str, columns={"ASA": "NOC"})
noc.head()
```

```python
m3 = pd.merge(oo, noc, on='NOC')
m3.head()
```

```python
m3.tail()
```

```python
m4 = pd.merge(oo, noc, on='NOC',how='left')
m4.head()
```

```python
m4.tail()
```

```python
m5 = pd.merge(oo, noc, on='NOC',how='right')
m5.head()
```

```python
m5.tail()
```
