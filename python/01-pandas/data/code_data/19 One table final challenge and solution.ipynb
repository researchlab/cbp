```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oo = pd.read_csv('data/olympics.csv',skiprows=4)
oo.head()
```

### In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.

```python
us = oo[oo.NOC == 'USA']
us.head()
```

```python
us  = us.groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value=0)
us
```

```python
us['Total'] = us.Gold + us.Silver + us.Bronze
us
```

```python
us.reset_index(inplace=True)
```

```python
us
```

```python
for year , group in us.groupby('Edition'):
    print(group.sort_values('Total',ascending=False)[:1])
```

```python
us_top = [ group.sort_values('Total',ascending=False)[:1] for year , group in us.groupby('Edition')]
```

```python
us_top  
```

```python
us_top_df = pd.DataFrame()
for i in us_top:
    us_top_df = us_top_df.append(i)
us_top_df    
```
