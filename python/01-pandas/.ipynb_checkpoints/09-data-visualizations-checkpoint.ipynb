# Data visualizations

## Seaborn with heatmaps

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
oo = pd.read_csv('data/olympics.csv',skiprows=4)
oo.head()
```

## Using the Olympic dataset, present a summary of the total medals won by participating countries in the 2008 Olympics.

```python
c = oo[oo.Edition == 2008]
c.head()
```

```python
g = c.groupby(['NOC','Medal']).size()
```

```python
g= g.unstack('Medal',fill_value=0)
g
```

```python
g = g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
g.head()
```

```python
sns.heatmap(g)
```

```python
g = g.transpose()
g
```

```python
plt.figure(figsize=[16,5])
sns.heatmap(g)
```

# Creating your own colormaps

```python
g = oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head().plot(kind='bar')
```

```python
from matplotlib.colors import ListedColormap
```

```python
sns.color_palette()
```

```python
sns.palplot(sns.color_palette())
```

```python
gsb = ['#dbb40c','#c5c9c7','#a87900']
sns.palplot(sns.color_palette(gsb))
```

```python
my_gsb = ListedColormap(sns.color_palette(gsb))
```

```python
g = oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g = g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
```

```python
g.head().plot(kind='bar',colormap=my_gsb)
```

## In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.

```python
us = oo[oo.NOC == 'USA']
us.head()
```

```python
us = us.groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value=0)
us
```

```python
us['Total'] = us.Gold + us.Silver + us.Bronze
us
```

```python
us.reset_index(inplace=True)
us
```

```python
for year, group in us.groupby('Edition'):
    print(group.sort_values('Total',ascending=False)[:1])
```

```python
us_top = [group.sort_values('Total',ascending=False)[:1] for year, group in us.groupby('Edition')]
us_top
```

```python
us_top_df = pd.DataFrame()
for i in us_top:
    us_top_df = us_top_df.append(i)
us_top_df
```
