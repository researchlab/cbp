```python
import pandas as pd
```

```python
oo = pd.read_csv('data/olympics.csv',skiprows=4)
oo.head()
```

# Athlete winning medals in Beijing Olympics 100m or 200m track event.

```python
mw = oo[(oo.Edition == 2008) & ((oo.Event == '100m')|(oo.Event == '200m'))]
mw.head()
```

```python
mw.tail()
```

```python
g = mw.groupby(['NOC','Gender','Discipline','Event']).size()
g
```

```python
df = g.unstack(['Discipline','Event'])
df
```

## stack() 把行变成列

```python
df.stack()
```

```python
df.stack('Event')
```

## unstack() 把列变成行

```python
df
```

```python
df.unstack()
```

```python
df.unstack('Gender')
```

# chanenge

Plot the number of gold medals won by the USA male and female athletes throughout the history of the Olympics?

```python
am = oo[(oo.NOC == 'USA') & (oo.Medal == 'Gold')]
am.head()
```

```python
am.groupby(['Edition','Gender']).size().plot(kind='bar')
```

```python
gu = am.groupby(['Edition','Gender']).size().unstack('Gender', fill_value=0)
```

```python
gu.plot()
```

Plot the 5 athletes who have won the most gold medals over the history of the Olympics. When there is a tie, consider the number of silver medals, then bronze medals.

```python
g = oo.groupby(['Athlete','Medal']).size()
g.head()
```

```python
g = g.unstack('Medal',fill_value=0)
```

```python
g = g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
g.head()
```

```python
g.head().plot(kind='bar')
```
